```python
import requests
stamen = folium.Map(location=[45.7640, 4.8357], zoom_start=13)

r = requests.get('http://api.citybik.es/velov.json')
data = r.json()

for i in data:
    #Need to change formatting: from GPS coordinates expressed in integers, to float. 
    if (i['free'] > (i['bikes']*0.4)): #show green marker for stations with more than 40% available spots
        x = float(i['lat']/1000000)
        y = float(i['lng']/1000000)
        stamen.simple_marker([x, y], popup=i['name'], marker_color='green', marker_icon='ok-sign')
    else:
        x = float(i['lat']/1000000)
        y = float(i['lng']/1000000)
        stamen.simple_marker([x, y], popup=i['name'],marker_color='red', marker_icon='remove-sign')
        
stamen
# 346 bike stations in Lyon
```
